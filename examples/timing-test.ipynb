### validate
check that `py_ts_time + ts_py_time = ts_roundtrip` 
and that there are the same number entries in a lists (may need to wait a second for all messages to finish)

```python
import numpy as np
assert np.all(np.array(w.py_ts_times) + np.array(w.ts_py_times) == np.array(w.ts_roundtrip))
#check that they all have the same length
len(w.py_ts_times), len(w.ts_py_times), len(w.ts_roundtrip)
```

```python
plt.title('roundtrip time')
# plt.hist([i[0]+i[1] for i in zip(w.py_ts_times, w.ts_py_times)], bins='auto', density=True)

plt.hist(w.ts_roundtrip, bins='auto', density=True)
ax = plt.gca()
xlim = ax.get_xlim()
plt.xlabel('ms')
plt.tight_layout()
plt.savefig('roundtrip.png')
plt.show()

# idx = int(np.max(w.py_ts_times) > np.max(w.ts_py_times))
# bins = np.histogram_bin_edges([w.ts_py_times,w.py_ts_times][idx],bins='auto')

plt.hist(w.ts_py_times, bins = 'auto', density=True)
plt.xlim(xlim)
plt.title('typescript -> python')
plt.xlabel('milliseconds')
plt.tight_layout()
plt.savefig('ts_to_py.png')
# plt.ylabel('probability')
plt.show()

plt.hist(w.py_ts_times, bins= 'auto', density=True)
plt.xlim(xlim)
plt.title('python -> typescript')
plt.xlabel('milliseconds')
plt.tight_layout()
plt.savefig('py_to_ts.png')

# plt.ylabel('probability')
plt.show()





plt.scatter(w.ts_py_times,w.py_ts_times, alpha=.1, color='k')
plt.ylabel('py -> ts (ms)')
plt.xlabel('ts -> py (ms)')
plt.tight_layout()
plt.savefig('scatter.png')
plt.show()
```
